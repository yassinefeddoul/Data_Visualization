# Visualisation avec Matplotlib

```python
 import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
import matplotlib.pyplot as plt
# pour les utilisateur de jupyter notebook
%matplotlib inline
```

# Question 1

```python
plt.plot(x, y)
plt.xlabel('x')
plt.title('title')
```

# Question 2

```python
figure = plt.figure()
ax1 = figure.add_axes([0, 0, 1, 1])
ax2 = figure.add_axes([0.2, 0.5, .4, .4])
ax2.set_xlim(20, 22)
ax2.plot(x, y)
ax1.plot(x, z)
ax1.set_xlabel('x')
ax2.set_xlabel('x')
ax1.set_ylabel('Z')
ax2.set_ylabel('Y')
ax2.set_title('Y en fonction de X')
ax1.set_title('Z en fonction de X')
```

# Question 3

```python
fig, axs = plt.subplots(1, 2)
ax1 = plt.gca()
ax1.set_xlabel("x")
ax1.set_ylabel("z")
fig.set_size_inches(10, 7)
axs[0].plot(x, y, "b--")
axs[1].plot(x, z,"r")
```
